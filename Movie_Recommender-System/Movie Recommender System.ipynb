```python
import ast   #Abstract Syntax Trees
```

__Literal - A datatype that can hold any value types such as strings, numbers and more.__

__ast.literal_eval - To safely evaluate strings containing python values from unknown sources without us having to parse the values.__

__Parsing - Method of translating code into machine language to investigate the appropriate syntax of the code.__

```python
movies['genres'][0]
```

```python
def conversion(x):
    l = []
    for i in ast.literal_eval(x):
        l.append(i['name'])
    return l
```

```python
movies['genres'] = movies['genres'].apply(conversion)
```

```python
movies['genres']
```

```python
movies.head()
```

```python
movies['keywords'][0]
```

```python
movies['keywords'] = movies['keywords'].apply(conversion)
```

```python
movies.head()
```

```python
movies['cast'][0]
```

```python
# To get 3 actor names per movie:
def conv(y):
    L = []
    count = 0
    for i in ast.literal_eval(y):
        if count !=3:
            L.append(i['name'])
            count+=1
        else:
             break
    
    return L
```

```python
movies['cast'] = movies['cast'].apply(conv)
```

```python
movies['cast']
```

```python
movies.head()
```

```python
def director(d):
    l = []
    for i in ast.literal_eval(d):
        if i['job'] == 'Director':
            l.append(i['name'])
            break
    return l
```

```python
movies['crew'] = movies['crew'].apply(director)
```

```python
movies['crew']
```

```python
movies.head()
```

```python
movies['overview'][0]
```

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split())
```

```python
movies['overview']
```

```python
movies.head()
```

```python
# Removing unnecessary spaces to avoid misconception:
movies['genres'] = movies['genres'].apply(lambda x: [i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x: [i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x: [i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

# Stemming:

```python
import nltk
from nltk.stem.porter import PorterStemmer
```

```python
porterstem = PorterStemmer()
```

```python
def stemming(text):
    l = []
    for i in text.split():
        l.append(porterstem.stem(i))
        
    return " ".join(l)
```

```python
df['tags'] = df['tags'].apply(stemming)
```

# Count Vectorizer:

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(df['tags']).toarray()
```

```python
vectors
```

```python
vectors[0]
```

```python
cv.get_feature_names_out()   #Most frequent words
```

# Cosine Similarity:

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors).shape
```

```python
similarity = cosine_similarity(vectors)
```

# Recommender System:

```python
def recommend(movie):
    movieIndex = df[df['title'] == movie].index[0]
    dist = similarity[movieIndex]
    moviesList = sorted(list(enumerate(dist)), reverse=True, key=lambda x: x[1])[1:6]
    
    for i in moviesList:
        print(df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```

```python
recommend('Batman Begins')
```
